# Step 1: Load the dataset

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

bank_data = pd.read_csv('bank-full.csv', sep=';')
```

# Step 2: Take a quick look at the data



```python
print("First few rows of the dataset:\n", bank_data.head())
```

# Step 3: Convert categorical data into numerical data


```python
bank_data_encoded = pd.get_dummies(bank_data, drop_first=True)
```

# Step 4: Define the input features (X) and the target variable (y)


```python
X = bank_data_encoded.drop('y_yes', axis=1)  
y = bank_data_encoded['y_yes']
```

# Step 5: Split the dataset into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Step 6: Create a Decision Tree Classifier


```python
clf = DecisionTreeClassifier(random_state=42)
```

# Step 7: Train the model


```python
clf.fit(X_train, y_train)
```

# Step 8: Make predictions on the test set

```python
y_pred = clf.predict(X_test)
```

# Step 9: Evaluate the model's performance



```python
accuracy = accuracy_score(y_test, y_pred)
print("\nAccuracy of the Decision Tree model:", accuracy)
```
