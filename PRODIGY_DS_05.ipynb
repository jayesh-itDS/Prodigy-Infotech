```python
#Converting the date time to datetime format
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d-%m-%Y %H:%M:%S')
df.set_index('DateTime', inplace=True)
#it is done to extract exact date time or year from the index at time of visualization
```

```python
df
```

```python
#VISUALIZING THE DATA
plt.figure(figsize=(10,8))
sns.countplot(x='Location Description', data=df)
plt.title('Accidents Hotspots')
plt.ylabel('Accidents')
plt.xlabel('Locations')
plt.show()
```

```python
plt.figure(figsize=(7,6))
sns.scatterplot(x='Longitude', y='Latitude', data=df, alpha=0.1)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Road Conditions', data=df)
plt.show()
```

```python
# Time of Day vs. Accident Count
plt.figure(figsize=(20, 12))
sns.countplot(x='Time', data=df)
plt.title('Accidents by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Accident Count')
plt.show()
```

```python
# Distribution of accidents based on month
plt.figure(figsize=(20, 12))
df.index.month.value_counts().sort_index().plot(kind='bar',color="purple")
plt.title('Distribution of Accidents based on Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.xticks(range(1, 13), ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'], rotation=45)
plt.show()
```

```python
# Distribution of accidents based on Time
plt.figure(figsize=(8, 6))
df.index.year.value_counts().sort_index().plot(kind='bar',color='orange')
plt.xlabel('Year')
plt.ylabel('Accidents')
plt.title('Distribution of accidents based on Time')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of accidents based on Time
plt.figure(figsize=(10, 6))
df.index.day.value_counts().sort_index().plot(kind='bar',color='red')
plt.xlabel('Year')
plt.ylabel('Accidents')
plt.title('Distribution of accidents based on Day')
plt.xticks(range(7), ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], rotation=45)
plt.show()
```

```python
#Distribution of accidents based on Weather
plt.figure(figsize=(10,6))
sns.countplot(x='Weather Conditions', data=df)
plt.xlabel('Weather Condition')
plt.ylabel('Accidents')
plt.title('Distribution of accidents based on Weather')
plt.show()
```
