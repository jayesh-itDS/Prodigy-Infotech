```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Age': [23, 25, 22, 30, 35, 30, 25, 22, 23, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male', 
               'Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male', 
               'Male', 'Female', 'Female', 'Male']
}

# Create a DataFrame
df = pd.DataFrame(data)

# 1. Bar Chart for Gender Distribution
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(10, 5))
plt.bar(gender_counts.index, gender_counts.values, color=['blue', 'orange'])
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Gender Distribution')
plt.xticks(rotation=45)
plt.show()

# 2. Histogram for Age Distribution
plt.figure(figsize=(10, 5))
plt.hist(df['Age'], bins=10, edgecolor='black', color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution Histogram')
plt.grid(axis='y')
plt.show()
```
